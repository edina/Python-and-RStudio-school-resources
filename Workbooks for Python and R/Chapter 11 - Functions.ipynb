## Introduction to Functions

<div class = "alert alert-block alert-success">

Functions are chunks of code that tell Python to do something. The great thing about functions is that you can reuse the code time and time again throughout your program without having to retype it all – this becomes very handy in more complicated programs.  
    
To write a function, you simply write the word ‘def’ (which stands for define) followed by a unique name for your function, 2 parentheses and a colon. **For example, myfunction( ):**  
    
    
After you have given your function a name, go to the next line and ensure your cursor is indented. Now, simply write the code as you usually would. This is the code that will run each time the function is called.  
    
In the program below, you will see 2 functions called **correct( )** and **incorrect( )**  
    
   
</div>

    

## Global variables

<div class = "alert alert-block alert-success">
    
You might also notice a new word being used near the top of the code – **global**. Let’s quickly explain what this means:  
    
Inside the correct( ) function, you will see that we write score = score + 1. That line of code simply says that if the user has got the question correct, then add a point to their current score.  
    
Because we are setting a new score inside a function, this score becomes known as a ‘local variable’ and only the code inside that function can see what the actual score is. This will cause issues when we try to access the score later in the program as the rest of the code can’t see what the true value of ‘score’ actually is.  
    
To get around this, we need to change the score variable to what’s known as a **global variable**. This means that the value of the variable ‘score’ can be seen throughout the entire program. You simply tell Python inside the function that you want the variable score to be global by writing global score.

```python
questions = [
     ['What is 1+2? ', ['a. 1', 'b. 2', 'c. 3'], ['c', '3']],
     ['What is the capital of Somalia? ', ['a. Abuja', 'b. Mogadishu', 'c. Kigali'], ['b', 'Mogadishu']],
     ['Which US state has the capital Montgomery? ', ['a. Alabama', 'b. Iowa', 'c. Delaware'], ['a', 'Alabama']],
]

def correct():
  global score
  score += 1
  print('Correct!')

def incorrect(answers):
  print('Incorrect. The answer is', ' or '.join(answers))

score = 0
for [question, answers, correct_answers] in questions:
    prompt = '\n{}\n{}\n'.format(question, '\n'.join(answers))
    answer = input(prompt)
    correct() if answer in correct_answers else incorrect(correct_answers)

messages = [
  'Embarassing.',
  'Better luck next time.',
  'Nearly full marks.',
  'Well done!',
]
print('\nYour score was {}.\n{}'.format(score, messages[score]))
```
