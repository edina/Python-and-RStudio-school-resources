## Introduction to Loops

<div class = "alert alert-block alert-success">

One of the most powerful things about computers is that they can repeat things over and over very quickly.
In most applications that you write, you’re going to need the ability to run the same code more than once. Rather than writing the same lines of code several times, we use loops. 

Loops will always run until a condition is met. For example, in a first-person shooter, the game will continue to
run until either you’ve won, or your health reaches zero. Once either of those conditions occur, the game ends.

There are 2 types of loops that we will look at in Python – While loops and For loops.      
</div>

    

## While Loops


<div class = "alert alert-block alert-success">
    
A while loop initially checks a condition and will then repeatedly execute a block of code while that condition is true. For example, you might want to write a program where the user has to continually guess what your favourite food is until they get it right. The code below does this:  
    
## While Loop example 1  
    
    guess = input("Guess what my favourite food is: ")  
    while guess!= "chocolate"  
    guess = input("Incoreect. Try again. Guess what my favourite food is: ")  
    print("Correct - my favourite food is chocolate.")

This code basically says that while the user has not guessed ‘chocolate’ as your favourite food, keep asking them to guess what your favourite food is. Once they get it right, we jump out of the loop, tell the user they guessed correctly, and the program will end.

       
 
</div>
    


<div class = "alert alert-block alert-success">
    
## While Loop Example 2
    
The code shown in the image below basically says that while the outside temperature is equal to or greater than 30 degrees, set the air conditioner to on. We continue monitoring the outside temperature (by repeatedly asking the user to enter temperature) and keep the air conditioner running while the outside temperature is 30 degree or higher. As soon as the temperature drops below 30 degrees, we exit 
out of the loop and set the air conditioner to off. The program will then end.

    temperature = int(input("Enter the outside temperature: "))  
    
    while temperature >=30:  
        print("Set air conditioner to ON")
        print()
        temperature = int(input("Enter the outside temperature: "))
    
    print("Set air conditioner to OFF")
       
</div>
    

<div class = "alert alert-block alert-success">
    
## While Loop Example 3
    
In a nutshell, this code counts to 50. It starts by setting a variable to 1. This will be a counter that will keep track of how many times we print a number on to the screen. On line 2, we reach the While loop. Here we can see that while the counter is less than or equal to 50, we are going to perform the indented code beneath it. The code inside of the loop will first of all print the value of the variable, ‘counter’. Each time we print out the value of the counter, we go ahead and add 1 to the counter. This code keeps repeating until the counter reaches 50. Once it reaches 50, the loop ends and the program will stop running as there is no further code to process.
      
    counter = 1
    while counter <=50:
        print(counter)
        counter = counter+1
                       
</div>

The computer generates this number by accessing the random library and running the randint (short for random integer) function from it. The brackets at the end of this line tell the computer the lower and upper parameters that it may pick a random number from – in this case, the computer will generate a number between 1 and 10.

## For Loops

<div class = "alert alert-block alert-success">
   
For loops are like While loops, in that they repeat a section of code. However, with For loops, we tell Python the exact number of times the code must be repeated (While loops keep running until a condition is met – they could even run forever!). For loops are generally used for iterating (repeating something) over a list of items. Let’s look at some examples:
    
 ## For Loops Example 1
    
To print hello five times in Python, you could do the following:

    print("Hello)  
    print("Hello")  
    print("Hello")  
    print("Hello")  
    print("Hello")
    print("Hello")
     
    
But this is rather tedious. Instead, use a For loop to reduce the amount of typing and repetition, like this:

    for i in range(0,5)
    print("Hello")
    
The code is telling Python to do the following:  
    
•	Start counting from 0 and stop before reaching 5  
•	For each number we count, store the value in the variable i  
•	Each time we count a number, Python executes the indented block of code on the second line  


    
</div>

<div class = "alert alert-block alert-success">
    
## For Loops Example 2
Python often uses For loops to iterate (loop) over a list of items and perform a certain function (we will look more at Lists in the next chapter). Take the example below, where we have a list of suburbs near Glasgow:  
    
    suburbs= ["East Kilbride", "Strathaven", "Eaglesham", "Newton Mearns"]
    print("Suburbs near Glasgow include:")  
    for i in suburbs:
        print("*" + i)
 
The For loop code tells Python to:  
    
•	Start the loop at East Kilbride in the suburbs list  
•	For each item that we iterate over in the list, store the value in variable i  
•	Each time we hit a new suburb in the list, run the indented code that prints out the name of the suburb
    
 </div>


```python
# Example for teacher use

suburbs= ["East Kilbride", "Strathaven", "Eaglesham", "Newton Mearns"]
print("Suburbs near Glasgow include:")  
for i in suburbs:
    print("*" + i)
```

# Range function  
   
Using the range function is helpful here. The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number. 
    
So in the syntax - **range(start, stop, step)**.  
    
    **Start** is the first number in your sequence.-  
    **Stop** is the last number.  
    **Step** is optional  - if you leave it out, the default is 1. You can specify an increment - in our example it will be 5.
